```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Matplotlib Figures, Subplots, Colors, Markers and Linestyle

## Creating Figures and Subplots

```python
array = np.arange(10,20)
array
```

```python
plt.plot(array)
```

```python
fig = plt.figure()
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(2,2,1) # (2,2) represents it contains 2 rows 2 columns and 1 represents this is 1 subplot
```

Note: We need to create figure and subplots inside one cell otherwise it does not work.

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(np.arange(1,6))
```

## Another way to create subplots

```python
fig,ax = plt.subplots(3,3,figsize = (8,6))
ax[0,0].plot([1,2,3,4,5])
ax[2,2].plot([1,2,3,4,5])
fig.subplots_adjust(hspace = .4 ,wspace = .4) #to put spacing between figure plots
```

# Changing colors, Markers and linestyle

```python
x = np.arange(0,200,10)
x
```

```python
y = np.random.randn(20).cumsum()
y
```

## Color

```python
plt.plot(x,y,color = 'r')
```

## Marker

```python
plt.plot(x,y,color = 'y',marker = 'o')
```

## Linestyle

```python
plt.plot(x,y,color = 'b',marker = 'v',linestyle = '--')
```
