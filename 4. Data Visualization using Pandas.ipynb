# Data Visualization using Pandas

## Plotting series

```python
series1 = pd.Series(np.random.randn(10).cumsum())
series1
```

```python
series1.plot()
```

## Plotting dataframe

```python
data = pd.read_csv('data/ex7.csv',index_col='year')
data
```

```python
data.plot()
```

## Creating subplots

```python
data.plot(subplots = True,#to create subplots
          layout = (1,3), # layout of subplots
          figsize = (10,4), #to change the figure size
          legend = False, #to turn off the legneds
          title = 'Company Sales for Three Items', #to set title
          sharey = True) # to make y axis identical in all plots
```
