```python
import pandas as pd
import seaborn as sns
```

# Scatter and Pair Plots

Scatter plots are very useful to explore the relationship between two data series.

```python
data = pd.read_csv('data/boston.csv')
data.head()
```

```python
sns.regplot(x = 'medv', y ='crim', data  = data)
```

```python
sns.regplot(x = 'medv', y ='crim', data  = data,ci = False, fit_reg=False)
```

# Pairplot

```python
sns.pairplot(data)
```
